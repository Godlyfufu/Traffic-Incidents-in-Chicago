```python
import pandas as pd

# Load your dataset
data = pd.read_csv(r"C:\Users\matth\0_BDMP\Datasets\1_Crash_Datasets\Speed_Camera_Locations.csv")
data.head()
```

```python

# Extract the first four numbers from the 'Address' column
data['ADDRESS'] = data['ADDRESS'].str.extract(r'^(\d{3,5})')

# Print the modified 'Address' column
print(data['ADDRESS'])
```

```python
data.head(30)
```

```python
# Create a dictionary to specify the column name mappings
column_names = {'LATITUDE': 'speed_camera_latitude',
                'LONGITUDE': 'speed_camera_longitude',
                'LOCATION': 'speed_camera_location'}

# Rename the columns using the dictionary
data = data.rename(columns=column_names)

# Print the updated column names
print(data.columns)
```
