```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# File paths for the datasets
crash_data_file = r"C:\Users\matth\0_BDMP\Datasets\1_Crash_Datasets\Traffic_Crashes_-_Crashes_damage.csv"
vehicles_data_file = r"C:\Users\matth\0_BDMP\Datasets\1_Crash_Datasets\Traffic_Crashes_-_People.csv"
people_data_file = r"C:\Users\matth\0_BDMP\Datasets\1_Crash_Datasets\Traffic_Crashes_-_Vehicles.csv"

# Read the datasets into Pandas DataFrames
crash_data = pd.read_csv(crash_data_file)
vehicles_data = pd.read_csv(vehicles_data_file)
people_data = pd.read_csv(people_data_file)

# Merge the datasets based on the common key (CRASH_RECORD_ID)
merged_data = pd.merge(crash_data, vehicles_data, on="CRASH_RECORD_ID", how="inner")
merged_data = pd.merge(merged_data, people_data, on="CRASH_RECORD_ID", how="inner")

# Perform feature analysis
feature_analysis = merged_data.describe(include="all")  # Statistical summary of all features
correlation_matrix = merged_data.corr()  # Correlation matrix of numeric features

# Print the results
print("Feature Analysis:")
print(feature_analysis)
print("\nCorrelation Matrix:")
print(correlation_matrix)
```
