```python
# Checking unique values in categorical columns
for column in df.columns:
    if df[column].dtype == 'object':
        print(f"Column: {column}")
        print(df[column].unique())
        print("\n")
```

# Data Transformation

## Binning Time

```python
# observing values of crash_hour
df.CRASH_HOUR.value_counts()
```

```python
df['TIME_BINS'] = pd.cut(x=df['CRASH_HOUR'], bins = [0,6,12,18,23], 
                         labels = ['LATE NIGHT', 
                        'MORNING', 'AFTERNOON','EVENING'])
df.head(10)
```

## Binning Age

```python
df.AGE.value_counts()
```

```python
# creating bins and labels, preview data
df['AGE_BINS'] = pd.cut(x=df['AGE'], bins = [0,15,24,35,55,100], 
                         labels = ['< 15', '16-24', 
                        '25-35', '36-55','> 55'])
df.head()
```

## Binning Speed

```python
df.POSTED_SPEED_LIMIT.value_counts()
```

```python
# creating bins and label, previewing data 
df['SPEED_BINS'] = pd.cut(x=df['POSTED_SPEED_LIMIT'], bins = [0,15,25,40,75], 
                         labels = ['0-15', '16-25', 
                        '26-40', '> 41'])
df.head()
```

## Binning Injury Severity

```python
df.info()
```

```python
unique_most_severe_injury_values = df['MOST_SEVERE_INJURY'].unique()
print(unique_most_severe_injury_values)
```

```python
classification_mapping = {
    'REPORTED, NOT EVIDENT': 0,
    'NO INDICATION OF INJURY': 0,
    'NONINCAPACITATING INJURY': 1,
    'INCAPACITATING INJURY': 2,
    'FATAL': 3
}

df['MOST_SEVERE_INJURY'] = df['MOST_SEVERE_INJURY'].map(classification_mapping)
```
